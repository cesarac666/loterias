# Frequencia das dezenas (BOLA DA VEZ)

```python
def calcular_sequencia_atraso_atual(df_resultados, numero):
    # Certificando-se de que os resultados estão ordenados do mais recente para o mais antigo
    df_ordenado = df_resultados.sort_values(by='CC', ascending=False)

    sequencia_atual = 0
    atraso_atual = 0
    encontrado = False

    # Iterando sobre os resultados do mais recente para o mais antigo
    for index, row in df_ordenado.iterrows():
        if numero in row.values:
            if not encontrado:
                encontrado = True
            sequencia_atual += 1
        else:
            if not encontrado:
                atraso_atual += 1
            else:
                break  # Interrompe a contagem de atraso assim que o número for encontrado

    return sequencia_atual, atraso_atual

# Exemplo de uso
numero = 1  # Substitua pelo número desejado
sequencia_atual, atraso_atual = calcular_sequencia_atraso_atual(dfresultados, numero)
print(f"Sequencia Atual para o número {numero}: {sequencia_atual}")
print(f"Atraso Atual para o número {numero}: {atraso_atual}")
```

```python
def calcular_maior_sequencia_atraso(df_resultados, numero):
    maior_sequencia = 0
    maior_atraso = 0
    sequencia_atual = 0
    atraso_atual = 0

    for index, row in df_resultados.iterrows():
        if numero in row.values:
            sequencia_atual += 1
            atraso_atual = 0
        else:
            atraso_atual += 1
            sequencia_atual = 0

        if sequencia_atual > maior_sequencia:
            maior_sequencia = sequencia_atual
        if atraso_atual > maior_atraso:
            maior_atraso = atraso_atual

    return maior_sequencia, maior_atraso
```

```python
# Calculando a frequência de cada número
frequencia_numeros = {i: 0 for i in range(1, 26)}  # Inicializa um dicionário para armazenar as frequências

for coluna in dfresultados[['B1', 'B2', 'B3', 'B4', 'B5', 'B6', 'B7', 'B8', 'B9', 'B10', 'B11', 'B12', 'B13', 'B14', 'B15']].columns:
    for numero in range(1, 26):
        frequencia_numeros[numero] += dfresultados[coluna].eq(numero).sum()

# Convertendo o dicionário de frequência em um DataFrame para melhor visualização
df_frequencia = pd.DataFrame(list(frequencia_numeros.items()), columns=['Numero', 'Frequencia'])
df_frequencia.sort_values(by='Frequencia', ascending=False, inplace=True)

maiores_sequencias_atrasos = [calcular_maior_sequencia_atraso(dfresultados, numero) for numero in df_frequencia['Numero']]
df_frequencia['MaiorSequencia'], df_frequencia['MaiorAtraso'] = zip(*maiores_sequencias_atrasos)

sequencias_atrasos = [calcular_sequencia_atraso_atual(dfresultados, numero) for numero in df_frequencia['Numero']]
df_frequencia['SequenciaAtual'], df_frequencia['AtrasoAtual'] = zip(*sequencias_atrasos)
```

```python
def maior_sequencia_consecutiva(row):
    # Extrai apenas os números das colunas B1 a B15
    numeros = sorted([row[f'B{j}'] for j in range(1, 16)])
    maior_sequencia = 0
    sequencia_atual = 1

    # Itera pelos números para contar a maior sequência consecutiva
    for i in range(1, len(numeros)):
        if numeros[i] == numeros[i - 1] + 1:
            sequencia_atual += 1
            maior_sequencia = max(maior_sequencia, sequencia_atual)
        else:
            sequencia_atual = 1

    return maior_sequencia

def adicionar_maior_sequencia_consecutivos(df):
    # Cria uma nova coluna 'countCS' aplicando a função a cada linha
    df['countCS'] = df.apply(maior_sequencia_consecutiva, axis=1)
    return df
```

```python
def adicionar_maior_salto(df):
    def calcular_maior_salto(numeros):
        # Ordena os números e adiciona o limite superior (26), para considerar o salto após o último número.
        numeros_ordenados = sorted(numeros) + [26]
        maior_salto = max(numeros_ordenados[i+1] - numeros_ordenados[i] for i in range(len(numeros_ordenados) - 1)) # - 1
        return maior_salto

    salto_coluna = []
    for _, row in df.iterrows():
        numeros = [row[f'B{j}'] for j in range(1, 16)]
        salto_coluna.append(calcular_maior_salto(numeros))

    df['maiorSalto'] = salto_coluna
    return df
```

```python
import builtins

def filtrar_nahval(dfApostas, ultimo_resultado, NAHs_validos):
    rows = []
    for index, row in dfApostas.iterrows():
        numbers = {row[f'B{j}'] for j in range(1, 16)}
        
        # Convertendo as strings para conjuntos de inteiros
        NS_to_N = set(map(int, ultimo_resultado['NS_to_N'].iloc[0].strip("[]").split(", ")))
        N_to_A = set(map(int, ultimo_resultado['N_to_A'].iloc[0].strip("[]").split(", ")))
        AH_to_H = set(map(int, ultimo_resultado['AH_to_H'].iloc[0].strip("[]").split(", ")))

        #print(f"Aposta {index}: NS_to_N={NS_to_N}, N_to_A={N_to_A}, AH_to_H={AH_to_H}")

        for nah in NAHs_validos:
            n, a, h = nah

            # countN = sum(1 for num in numbers if num in NS_to_N)
            countN = builtins.sum([1 for num in numbers if num in NS_to_N])

            countA = builtins.sum([1 for num in numbers if num in N_to_A])
            countH = builtins.sum([1 for num in numbers if num in AH_to_H])

            #print(f"Aposta {index}: countN={countN}, countA={countA}, countH={countH}")

            if countN == n and countA == a and countH == h:
                rows.append(row)
                #print("Aposta válida:", row)

    return pd.DataFrame(rows) 
```
